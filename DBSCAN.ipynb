```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
data = files.upload()
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs.head()
```

```python
moons = pd.read_csv('cluster_moons.csv')
moons.head()
```

```python
circles = pd.read_csv('cluster_circles1.csv')
circles.head()
```

```python
fig, axes = plt.subplots(figsize=(21, 5), nrows=1, ncols=3)
sns.scatterplot(x='X1', y='X2', data=blobs, ax=axes[0], color='purple')
sns.scatterplot(x='X1', y='X2', data=moons, ax=axes[1], color='purple')
sns.scatterplot(x='X1', y='X2', data=circles, ax=axes[2], color='purple')
```

```python
def display_categories(model, data):
  label = model.fit_predict(data)
  sns.scatterplot(x='X1', y='X2', data=data, hue=label, palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
display_categories(model, blobs)
```

```python
model = KMeans(n_clusters=2)
display_categories(model, moons)
```

```python
model = KMeans(n_clusters=2)
display_categories(model, circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
display_categories(model, blobs)
```

```python
model = DBSCAN(eps=0.15)
display_categories(model, moons)
```

```python
model = DBSCAN(eps=0.15)
display_categories(model, circles)
```
